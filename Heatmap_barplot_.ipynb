```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv", index_col=0)
df.columns=['maydon', 'aholi']
df
```

Yuqoridagi jadvalda O'zbekiston tarkibidagi hududlarning maydoni va aholi soni jamlangan. Keling ularni ustunli grafik yordamida solishtiramiz.

seabornda ustunli diagramma chizish uchun .barplot() metodiga murojaat qilamiz.

```python
plt.figure(figsize=(14,5))
sns.barplot(x=df.index, y=df['aholi'])
plt.show()
```

```python
plt.figure(figsize=(14,5))
sns.barplot(x=df.index, y=df['aholi'])
plt.title('Uzbekistan aholisi')
plt.xlabel('Hudud')
plt.ylabel('Aholi soni (million)')
plt.xticks(rotation=90)
plt.grid(color='g', alpha=0.7, linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(14,5))
sns.barplot(x=df.index, y=df['maydon'])
plt.title('Uzbekistan aholisi')
plt.xlabel('Hudud')
plt.ylabel('Hudud kv.km')
plt.xticks(rotation=90)
plt.grid(color='g', alpha=0.7, linestyle='--')
plt.show()
```

```python
df_m['year']=[i.year for i in df_m.index]
df_m
```

```python
bon = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")
bon
```

```python
bon.shape
```

Yuqoridagi jadval turli oylarda turli aviakompaniyalarga tegishli samolyotlar o'rtacha necha minutga kechikkani ko'rsatilgan***** (har bir ustun bu aviakompaniya nomi qisqartmasi).

*manfiy qiymat samolyot erta kelganini anglatadi.

Biz alohida aviakompaniya uchun oyma-oy qiymatlarni ustunli grafik ko'rinishida chiqarishimiz mumkin:

```python
plt.figure(figsize=(16,6))
sns.barplot(x=bon['Month'], y=bon['AA'])
plt.title("American Airlines kompaniyasi samolyotlari o'rtacha kechikish vaqti")
plt.ylabel('minuts')
plt.show()
```

Keling endi bir nechta aviakompaniyalarni yonma-yon solishtiramiz. Buning uchun jadvalimizni shakl o'zgartirishimiz kerak.

Pandasda df shaklini o'zgratirish uchun reshape yoki pd.melt() funksiyasidan foydalanishimiz mumkin.

melt funksiyasi jadvlani eniga uzundan bo'yiga uzun holatga olib kelishda qo'llaniladi. Bunda df dagi bir nechta ustunlarni yagona variable ustuniga keltirish mumkin. Qiymatlar esa har bir variable uchun value ustunida ko'rsatiladi.

Tushunarli bo'lishi uchun quyidagi misolga ahamiyat bering. Bu misolda biz AA va MQ aviakompaniyalari uchun ma'lumotlarni ajratib olganmiz.

```python
go=pd.melt(bon, id_vars=['Month'], value_vars=['AA', 'AS'])
go

```

```python

sns.catplot(data=go, kind='bar', x='Month', y='value', hue='variable')
plt.show()
```

```python
data=pd.melt(bon, id_vars=['Month'], value_vars=['AA','AS','DL','WN','MQ'] )
data.head()
```

```python
sns.catplot(data=data, kind='bar', x='Month', y='value', hue='variable')
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(data=bon.set_index('Month'), cmap='Greens', annot=True)
plt.show()
```

```python
plt.figure(figsize=(16,7))
sns.heatmap(data=bon.set_index('Month'), cmap='Blues', square=True, annot=True)
plt.xlabel('Airline')
plt.show()
```
