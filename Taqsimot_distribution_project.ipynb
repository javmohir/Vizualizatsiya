```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns  # agar seaborn o'rnatilmagan bo'lsa pip install seaborn
```

 VIZUALIZASIYA
 Distribution - taqsimot
Taqsimot grafik - o'zgaruvchi (ustun)da qanday qiymatlar mavjudligi va ularning ehtimolligini ko'rsatuvchi grafik.

Ushbu amaliyotda uybor.uz saytidan yuklab olingan ma'lumotlar tahlili bilan shu'gullanamiz.

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

```python
sns.histplot(data=df, x='price')
plt.show()
```

```python
df=df[df.price<200000]
sns.histplot(data=df, x='price')
plt.show()
```

```python
sns.histplot(data=df, x='size')
plt.show()
```

```python
sns.histplot(data=df, x='rooms')
plt.show()
```

```python
df=df[df['rooms']<6]
```

```python
sns.displot(data=df, x='price', hue='rooms', palette='Accent', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', hue='rooms', palette='Accent', height=5, aspect=1.5)
plt.show()
```

```python
sns.displot(data=df[df.rooms<4], x='price', hue='rooms', multiple='stack', palette='Reds', height=5, aspect=1.5)
plt.show()
```

```python
sns.displot(data=df[df['rooms']>3], x='price', col='rooms', height=5, aspect=1.5)
plt.show()
```

```python
sns.displot(data=df, x='price', kind='kde', height=5, aspect=1.5)
plt.show()
```

```python
sns.displot(data=df, x='price', kde='ecdf', height=5, aspect=1.5)
plt.show()
```

```python
sns.kdeplot(data=df[df['rooms'].isin([2,3,4])], x='price', hue='rooms', shade=True)
plt.xticks(rotation=90)
plt.xticks()
plt.show()
```

```python
sns.displot(data=df, x='size', kind='ecdf',  height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='price', kind='ecdf',  height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='major') # setka (major, minor, both) qiymatlarini berib ko'ring
plt.minorticks_on()
plt.show()
```
