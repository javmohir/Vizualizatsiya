```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns  # agar seaborn o'rnatilmagan bo'lsa pip install seaborn
```

VIZUALIZASIYA
AMALIYOT. scatterplot

Ushbu amaliyotda uybor.uz saytidan yuklab olingan ma'lumotlar tahlili bilan shu'gullanamiz.

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

 jadvalda Toshkent shahridagi sotildagian uylar haqida ma'lumotlar jamlangan.
 Uylarning narxi (price) va maydoni (size) o'rtasida bog'liqlikni ko'rsatuvchi scatter plot chizing.

```python
sns.scatterplot(data=df, x=df['price'], y=df['size'])
plt.show()
```

Jadvalda 7000 dan oshiq uy bor, aksar uylarning maydoni 250kv.m dan kam. Keling noodatiy katta qiymatlarni tashlab yuboramiz va yuqoridagi chizmani qaytadan chizamiz.

```python
df=df[(df['size']<250)&(df['price']<50000)]
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x=df['size'], y=df['price'])
plt.show()
```

Uylarning narxi (price) va maydoni (size) o'rtasida bog'liqlikni ko'rsatuvchi scatter va to'g'ri chiziqni chizish.

```python
df=df[(df['size']<250)&(df['price']<50000)]
plt.figure(figsize=(12,6))
sns.regplot(data=df, x=df['size'], y=df['price'])
plt.show()
```

Uylarning narxi, maydoni va xonalar soni o'rtasidagi bog'liqlikni ko'rsatish.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='size', y='price', hue='rooms')
plt.show()
```

Uylarning narxi, maydoni va tumanlar o'rtasidagi bog'liqlikni ko'rsatish.

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x='price', y='size', hue=df['district'])
plt.show()
```

```python
labels=['chilozor','yakkasaroy', 'uchtepa','mirzo ulugbek','olmazor','yashnobod','yunusobod','sergeli','shayxontohur','yangihayot','mirobod','bektemir' ]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='size', y='price', hue='district')
plt.legend(labels, loc=2)
plt.show()
```

Swarm ploit yordamida turli tumanlardag uylarning narxini taqqoslash.

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=df, x='district', y='price')
plt.xticks(rotation=90)
plt.show()
```
