```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns  # agar seaborn o'rnatilmagan bo'lsa pip install seaborn
```

VIZUALIZASIYA
Sublpot
subplot yordamida 1 ta chizmada bir nechta grafiklarni chizishimiz mumkin.

```python
x=np.arange(0, 4 * np.pi, 0.1)
sin=np.sin(x)
cos=np.cos(x)
plt.figure(figsize=(12,6))
plt.subplot(2,1,1) # 2 - jami qatorlar soni, 1 - jami ustunlar soni, 1-grafik
# sin(x) grafigi
plt.plot(x,  sin, linewidth=1.5, linestyle='-.', color='#8E44AD', alpha=0.7, marker='h')
plt.title('sin(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--')

plt.subplot(2,1,2) # 2 - jami qatorlar soni, 1 - jami ustunlar soni, 2-grafik
# cos(x) grafigi
plt.plot(x, cos, linewidth=1.5, linestyle=':', color='c', alpha=0.7, marker='v')
plt.title('cos(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--')

plt.show()
```

seaborn va subplot

```python
df=pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other']  # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other']  # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
df.head()
```

Bu safar subplots funksiyasidan foydalanamiz.
subplots funksiyasi 2 ta obyekt qaytaradi:
1. Umumiy chizma obyekti
2. Chizma tarkibidagi grafiklar ro'yxati
Biz quyida, qulaylik uchun, bu ikkalasini fig (figure) va ax (axes) o'zgaruvchilariga saqlab olamiz:

```python
sns.histplot(data=df, x='price')
plt.show()
```

```python
sns.scatterplot(data=df, x='price', y='year', s=100)
plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,10))
fig.suptitle('Grafig 1')
sns.histplot(ax=ax[0], data=df, x='price')
sns.scatterplot(ax=ax[1], data=df, y='price', x='year', hue='fuelType', s=100)
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(15,10))
# Umumiy chizma nomini beramiz:
fig.suptitle('Grafig 2')
sns.histplot(ax=ax[0,0], data=df, x='price')
sns.histplot(ax=ax[0,1], data=df, x='mileage')
sns.scatterplot(ax=ax[1,0], data=df, x='year', y='price', hue='fuelType', s=50)
sns.scatterplot(ax=ax[1,1], data=df, x='year', y='mileage', hue='fuelType', s=50)
# Har bir grafik uchun nom:
ax[0,0].set_title("Price distribution")
ax[0,1].set_title("Mileage distribution")
ax[1,0].set_title("Price vs Year correlation")
ax[1,1].set_title("Mileage vs Year correlation")
plt.show()
```
