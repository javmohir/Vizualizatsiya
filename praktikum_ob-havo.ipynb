Ustunli diagramma turli kategoriyadagi qiymatlarni solishtirish uchun juda qulay.



```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("/content/toshkent.csv", index_col=0)
df.head()
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(data=df)
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("yil")
plt.show()
```

Grafiklar nomini to'g'rilang (T, Tm va hokazo matnlarni o'zbek tilida chiqaring.)

```python
labels=["Yillik o'rtacha harorat",
        "Yillik o'rtacha maksimal harorat",
        "Yillik o'rtacha minimal harorat",
        "Yillik o'rtacha shamol tezligi",
        "Yomg'irli kunlar soni",
        "Bo'ronli kunlar soni",
        "Tumanli kunlar soni",
        "To'fonli kunlar soni",
        "Do'l yoqqan kunlar"]
plt.figure(figsize=(16,6))
sns.lineplot(data=df, linewidth=2)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.xlabel("yil")
plt.legend(labels, loc=1)
plt.show()
```

Yuqoridagi grafikda turli ko'rsatkichlar aralashib ketdi. Keling ustun turiga qarab alohida grafiklar chizamiz.
 Haroratni ko'rsatuvchi ustunlarni alohida chiziqli grafik qilib chiqaring.

```python
labels=["Yillik o'rtacha harorat",
        "Yillik o'rtacha maksimal harorat",
        "Yillik o'rtacha minimal harorat"]
plt.figure(figsize=(12,6))
sns.lineplot(data=df[['T','TM','Tm']], linewidth=2)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.xlabel('yil')
plt.ylabel('harorat')
plt.legend(labels, loc=0)
plt.show()
```

```python
labels=["Yomg'irli kunlar soni",
        "Bo'ronli kunlar soni",
        "Tumanli kunlar soni",
        "To'fonli kunlar soni",
        "Do'l yoqqan kunlar soni"]
plt.figure(figsize=(16,6))
sns.lineplot(data=df[['RA','SN','TS','FG','TN']], linewidth=2)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.xlabel('kunlar')
plt.ylabel('yil')
plt.legend(labels, loc=2)
plt.grid(color='g', alpha=0.7)
plt.savefig('ob-havo')
plt.show()

```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df.index, y=df['V'])
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.ylabel("O'rtacha shamol tezligi (km/soat)")
plt.xlabel("Yillar")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(25,8))
sns.heatmap(data=df[['RA','SN','TS','FG','GR']].T, cmap='Blues', linewidth=1, square=True)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
yticks_labels=["Yomg'ir","Qor","Bo'ron","Tuman","Do'l"]
plt.yticks(np.arange(5)+0.5, labels=yticks_labels)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
#plt.savefig('onail')
plt.show()
```
