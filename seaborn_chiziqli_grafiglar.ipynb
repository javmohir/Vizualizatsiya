4.2. Chiziqli grafiklar. seaborn kutubxonasi
seaborn matplotlib ustiga qurilgan kutubxonalardan bo'lib, grafiklar bilan ishlashni yana osonlashtiradi va muhimi chizmalarga ko'zga yoqadigan qilib chiroy beradi.

seaborn to'liq dokumentasiya: https://seaborn.pydata.org/index.html

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns # agar seaborn o'rnatilmagan bo'lsa pip install seaborn
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

Yuqoridagi jadvalda AQSHdagi ba'zi muzeylarga turli sanalarda tashrif buyurganlar soni saqlangan.

```python
df.shape
```

seaborn da chiziqli grafik chizish uchun .lineplot() metodiga murojaat qilamiz.

```python
sns.lineplot(data=df)
plt.show()
```

Chiziqli grafiklarning asosiy vazifasi biror jarayon/hodisani vaqt o'tishi o'zgarishini (tendensiyasini) kuzatish

Yuqoridagi grafikdan ham biz turli muzeylarga tashrif buyurganlar sonini chiroyli va muhimi tushunarli ko'rinishda ko'rishimiz va turli xulosalar qilishimiz mumkin.

Misol uchun, muzeylar orasida Avila Adobe muzeyi juda mashur ekan. Firehouse Museumga esa 2014-yil 9-Yanvar kuni juda ko'p odam tashrig buyurgan (bu jadvaldagi xatolikka ham ishora qilishi mumkin)

Agar df ning barcha ustunlarini emas, tanlangan ustunlarning grafigini chizaman desangiz, buning ham imkoni bor:

```python
plt.plot(df)
plt.show()
```

Ahamiyat bering, seaborn chiziqlarning rangi, shakli qo'shimcha ma'lumotlarni avtomat qo'yib ketdi.

Solishtirish uchun, quyida to'g'ridan-to'g'ri matplotlib yordamida chizilgan grafikni ko'rishingiz mumkin.

Keling, avvalgi darsda o'rgangan bilimlarimiz bilan grafikni epaqaga keltiramiz:

```python
plt.figure(figsize=(14,5))
plt.title('muziylardagi tashriflar')
plt.xlabel('sana')
plt.grid(alpha=0.5, color='g')
plt.ylabel('mijozlar soni')
plt.xticks(rotation=90)
sns.lineplot(data=df,)
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.title('muziylardagi tashriflar')
plt.xlabel('sana')
plt.grid(alpha=0.5, color='g')
plt.ylabel('mijozlar soni')
plt.xticks(rotation=90)
sns.lineplot(data=df['Avila Adobe'], marker='o')
plt.show()
```
